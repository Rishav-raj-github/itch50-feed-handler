# ITCH 5.0 Zero-Copy Feed Handler Demo
This notebook demonstrates parsing NASDAQ ITCH 5.0 binary message streams. It simulates binary message generation and feeds it into the Rust-based parser, demonstrating zero-copy casting speeds.


```python
# Generate a mock binary ITCH 5.0 file
import struct

print("Generating mock ITCH 5.0 binary data...")
# Format 'A': Type(1B), Timestamp(8B), Ref(8B), Side(1B), Shares(4B), Stock(8B), Price(4B)
timestamp = 34200000000000
order_ref = 1002594
buy_sell = b'B'
shares = 500
stock = b"AAPL    "
price = 1824500

payload = struct.pack(">QQsI8sI", timestamp, order_ref, buy_sell, shares, stock, price)
# Replace first byte with message type character (struct packs side as char but type needs custom prefix)
message = b'A' + payload[1:]

with open("mock_feed.bin", "wb") as f:
    f.write(message)
print("Saved mock_feed.bin. Ready for parsing.")

```

```python
# Compile and run the Rust parser against the binary feed
import subprocess

print("Compiling Rust feed handler...")
subprocess.run(["cargo", "build", "--release"], check=True)

print("Running parser benchmark...")
result = subprocess.run(["cargo", "run", "--release"], capture_output=True, text=True)
print(result.stdout)

```

The benchmark output above shows the performance metrics for zero-copy binary parser casting in nanoseconds.
