# Message Frequency & Burst Analysis
This notebook analyzes ITCH 5.0 message rates to identify message frequency spikes (bursted volume) and evaluates queue capacities to prevent packet drops.


```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
# Simulate 1000 intervals of message counts per millisecond
msg_counts = np.random.poisson(lam=12, size=1000)

# Simulate a network burst at interval 400
msg_counts[400:415] = msg_counts[400:415] * 8

# Queue capacity limit check
queue_limit = 80
overflows = np.where(msg_counts > queue_limit)[0]

print(f"Total simulated intervals: {len(msg_counts)}")
print(f"Max message rate/ms: {np.max(msg_counts)}")
print(f"Intervals exceeding queue limits: {len(overflows)}")

```

```python
plt.figure(figsize=(10, 4))
plt.plot(msg_counts, color="cyan", label="Message Rate (msg/ms)")
plt.axhline(queue_limit, color="red", linestyle="--", label="Queue Limit Threshold")
plt.title("Intraday Message Rate Feed Bursts")
plt.xlabel("Time (ms)")
plt.ylabel("Messages")
plt.legend()
plt.show()

```
